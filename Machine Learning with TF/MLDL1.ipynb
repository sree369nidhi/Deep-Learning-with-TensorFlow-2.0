```
import numpy as np
import tensorflow as tf
```

```
print("Setting Eager mode...")
tf.enable_eager_execution()
tfe = tf.contrib.eager
```

```
a = tf.constant(2)
b = tf.constant(3)

print(a)
print(b)
```

```
c = a+b
d = a*b
print(c)
print(d)
```

```
a = tf.constant([[2., 1.],
                 [1., 0.]], dtype = tf.float32)
b = np.array([[3., 0.],
              [5., 1.]], dtype = np.float32)

print(a)
print(b)
```

```
c = a+b
d = tf.matmul(a,b)

print(c)
print(d)
```
