```
hello = tf.constant('Hello')
print(hello)
```

```
print(hello.numpy())
```

```
a = tf.constant(2)
b = tf.constant(3)
c = tf.constant(5)

add = tf.add(a,b)
sub = tf.subtract(a,b)
mul = tf.multiply(a,b)
div = tf.divide(a,b)

print(add.numpy(), sub.numpy(), mul.numpy(), div.numpy())
```

```
mean = tf.reduce_mean([a,b,c])
sumof = tf.reduce_sum([a,b,c])
print(mean.numpy(), sumof.numpy())
```

```
m1 = tf.constant([[1., 2.], [3., 4.]])
m2 = tf.constant([[5., 6.], [7., 8.]])

prod = tf.matmul(m1, m2)

prod.numpy()
```
